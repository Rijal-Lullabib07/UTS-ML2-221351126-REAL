# Business Understanding

## Project Domain

Domain: Education & Learning Analytics
Sub-Domain: Student Performance Prediction, Academic Early Warning Systems, and Intelligent Decision Support Systems

Sistem prediksi performa akademik mahasiswa merupakan bagian dari learning analytics, yaitu pendekatan yang menggabungkan teknologi, analisis data, dan ilmu pembelajaran untuk meningkatkan proses belajar-mengajar. Proyek ini fokus pada pengembangan sistem cerdas berbasis machine learning yang dapat memberikan prediksi kategori performa akademik mahasiswa (misalnya: "Lulus", "Remedial", "Tidak Lulus") berdasarkan parameter input seperti nilai tugas, nilai UTS, dan kehadiran. Hal ini mendukung peran institusi pendidikan dalam memberikan intervensi yang tepat waktu dan berbasis data terhadap mahasiswa yang berisiko rendah performanya.

Dengan semakin berkembangnya dunia pendidikan digital, kebutuhan terhadap sistem pendukung keputusan berbasis kecerdasan buatan (AI) menjadi semakin mendesak, khususnya dalam konteks pemantauan performa akademik secara real-time dan adaptif.



## Problem Statements

Permasalahan utama yang diangkat dalam proyek ini mencakup:

1.Minimnya Sistem Prediktif Otomatis dalam Pendidikan
Banyak institusi pendidikan, terutama di tingkat perguruan tinggi, belum mengadopsi sistem berbasis AI untuk memprediksi performa mahasiswa. Evaluasi akademik umumnya dilakukan setelah ujian akhir semester, sehingga tidak ada sistem yang memberikan early warning sebelum terlambat.

2.Sulitnya Intervensi Dini terhadap Mahasiswa Berisiko Rendah
Tanpa prediksi performa, mahasiswa yang sedang mengalami penurunan akademik tidak terdeteksi lebih awal. Hal ini menyebabkan pihak pengajar atau pembimbing akademik kehilangan kesempatan untuk memberikan dukungan atau intervensi yang diperlukan tepat waktu.

3.Penilaian Akademik yang Cenderung Subjektif dan Tidak Konsisten
Dalam praktiknya, beberapa keputusan mengenai kemajuan akademik mahasiswa masih dilakukan secara subjektif oleh dosen, yang bisa menimbulkan ketidakkonsistenan dan ketidakadilan dalam evaluasi.

4.Keterbatasan Tenaga Pengajar dalam Memantau Individu Mahasiswa
Dengan jumlah mahasiswa yang banyak, tidak realistis bagi dosen untuk melacak satu per satu perkembangan performa akademik mahasiswa. Maka dari itu, diperlukan sistem otomatis yang dapat membantu dosen mengidentifikasi mahasiswa yang perlu diperhatikan lebih lanjut.

5.Kurangnya Pemanfaatan Teknologi dalam Proses Pengambilan Keputusan Akademik
Meskipun data seperti kehadiran dan nilai sudah tersedia, belum banyak institusi yang mengintegrasikan data ini ke dalam sistem cerdas yang mampu melakukan analisis lanjutan dan memberikan rekomendasi yang actionable.



## Goals

Tujuan utama dari proyek ini adalah untuk:

1.Mengembangkan sistem prediksi performa akademik mahasiswa berbasis Artificial Neural Network (ANN) yang mampu mengklasifikasikan mahasiswa ke dalam kategori performa seperti “Lulus”, “Remedial”, atau “Tidak Lulus”.

2.Mendorong penggunaan teknologi AI untuk mendukung pengambilan keputusan berbasis data di lingkungan pendidikan, khususnya dalam memantau dan mengevaluasi kemajuan mahasiswa secara berkelanjutan.

3.Menyediakan antarmuka aplikasi web interaktif berbasis Streamlit, yang dapat diakses dengan mudah oleh dosen atau admin kampus untuk memasukkan data input dan mendapatkan hasil prediksi secara instan.

4.Mengoptimalkan model dengan konversi ke TensorFlow Lite, agar aplikasi dapat dijalankan di perangkat dengan sumber daya terbatas (seperti smartphone atau tablet), memungkinkan mobilitas dan aksesibilitas yang lebih luas, khususnya bagi institusi dengan infrastruktur teknologi terbatas.

5.Meningkatkan kualitas intervensi akademik dengan menyediakan prediksi awal yang akurat, sehingga tindakan seperti bimbingan akademik atau pemberian remedial dapat dilakukan lebih tepat sasaran.



## Solution Statements

Untuk mencapai tujuan tersebut, solusi yang ditawarkan dalam proyek ini adalah:

1.Penerapan Model Machine Learning: Artificial Neural Network (ANN)
Model ANN dipilih karena kemampuannya dalam menangani klasifikasi non-linear dan kompleks dari data multidimensional. Model ini dilatih menggunakan dataset akademik fiktif yang mencakup nilai tugas, nilai UTS, dan kehadiran. Model ANN diproses melalui pipeline preprocessing yang melibatkan normalisasi data dan encoding label target, dan diuji untuk mencapai akurasi prediksi yang tinggi (lebih dari 90%).

2.Pengembangan Aplikasi Web Menggunakan Streamlit
Aplikasi Streamlit dikembangkan sebagai antarmuka input-output sederhana yang memungkinkan pengguna memasukkan parameter akademik mahasiswa dan segera mendapatkan hasil klasifikasi performa. Desain UI difokuskan pada kemudahan penggunaan oleh pengguna non-teknis, seperti dosen atau tenaga administrasi akademik.

3.Konversi Model ke TensorFlow Lite (TFLite)
Untuk meningkatkan portabilitas dan efisiensi, model ANN yang telah dilatih dikonversi ke format TensorFlow Lite. Ini memungkinkan aplikasi dijalankan di perangkat dengan sumber daya komputasi terbatas tanpa kehilangan akurasi prediktif secara signifikan, mendukung implementasi di berbagai perangkat dan platform.

4.Integrasi Prediksi dalam Proses Keputusan Akademik
Sistem ini dapat diintegrasikan ke dalam sistem informasi akademik (SIAKAD) sebagai fitur tambahan untuk evaluasi mahasiswa, atau digunakan secara mandiri oleh tenaga pendidik untuk keperluan pemantauan dan intervensi.

5.Skalabilitas dan Potensi Pengembangan Lanjutan
Sistem ini dapat dikembangkan lebih lanjut untuk mengakomodasi lebih banyak fitur input seperti nilai kuis, aktivitas LMS, atau data psikometrik. Selain itu, klasifikasi dapat diperluas menjadi prediksi nilai akhir atau prediksi risiko drop-out.

# Data Preparation

```python
df['Performance'] = df['G3'].apply(lambda x: 'High' if x >= 15 else 'Medium' if x >= 10 else 'Low')
```

```python
X = df.drop('Performance', axis=1)
y = df['Performance']
```

```python
for col in X.select_dtypes(include=['object', 'category']).columns:
    le = LabelEncoder()  # Create a new LabelEncoder for each column
    X[col] = le.fit_transform(X[col])
```

```python
# Scaling
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

# Modeling

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
print("Random Forest Accuracy:", rf.score(X_test, y_test))
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(3, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
from sklearn.utils import class_weight # Import the class_weight module

cw = class_weight.compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
cw_dict = dict(enumerate(cw))
```

```python
label_encoder = LabelEncoder()
```

```python
y_train_encoded = label_encoder.fit_transform(y_train)
```

```python
model.summary()
```

```python

from tensorflow.keras.utils import plot_model
plot_model(model, show_shapes = True)
```

```python
history = model.fit(X_train, y_train_encoded, epochs=50, batch_size=32, validation_split=0.2, class_weight=cw_dict, verbose=0)
```

# Evaluation

```python
y_test_encoded = label_encoder.transform(y_test)
loss, accuracy = model.evaluate(X_test, y_test_encoded, verbose=0)
print(f"Student Performance Accuracy: {accuracy * 100:.2f}%")
```

```python
y_pred_prob = model.predict(X_test)  # Get predicted probabilities
y_pred_classes = np.argmax(y_pred_prob, axis=1)  # Get predicted class indices

cm = confusion_matrix(y_test_encoded, y_pred_classes)  # Use predicted class indices

sns.heatmap(cm, annot=True, fmt='d', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(history.history["loss"], label='Training Loss', color='blue', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange', linestyle='--', linewidth=2)
plt.title('Model Loss Over Epochs', fontsize=14, fontweight='bold')
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend(fontsize=10, loc='upper right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
